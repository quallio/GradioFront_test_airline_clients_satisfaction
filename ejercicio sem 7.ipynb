```python
import pandas as pd
from funpymodeling.exploratory import status
```

```python
# PREPARACIÓN DE LOS DATOS
```

```python
# Cargamos la data. Viene separada en train y test, son dos files. Los mergeamos y luego separaremos si lo necesitamos.

data_train = pd.read_csv("Satisfaccion_del_vuelo/train.csv", sep=',', index_col=0)
data_test = pd.read_csv("Satisfaccion_del_vuelo/test.csv", sep=',', index_col=0)
all_data = pd.concat([data_train, data_test], axis=0)

# Uso solo estas filas para probar... porque sino queda muy pesado el modelo.
#all_data = pd.read_csv("Satisfaccion_del_vuelo/test.csv", sep=',', index_col=0)
```

```python
status(all_data)
```

```python
data = all_data[['Age', 'Class', 'Inflight wifi service', 'Ease of Online booking', 'Seat comfort', 'Checkin service', 'satisfaction']].copy()
```

```python
status(data)
```

```python
# Cambiamos los nombres de las columnas, para que no tengan espacios y sean más claros/concisos.

data.rename(
    columns = {
        'Inflight wifi service':'Wifi',
        'Ease of Online booking':'Booking',
        'Seat comfort':'Seat',
        'Checkin service':'Checkin',
        }, 
inplace = True)
```

```python
status(data)
```

```python
data
```

```python
# Cambiamos los valores de la columa "satisfaction":

class_map = {'neutral or dissatisfied':0, 'satisfied':1}
data['satisfaction'] = data['satisfaction'].map(class_map)
```

```python
data
```

```python
status(data)
```

```python
data1 = pd.get_dummies(data)
```

```python
status(data1)
```

```python
# Para averiguar los máx y mín de algunas columnas, para crear el front de Gradio

# La .T es de Transpuesta
data1.describe().T
```

```python
# CLASIFICACIÓN
```

```python
# Nuestro target o variable a clasificar es "satisfaction"

data_x = data1.drop('satisfaction', axis=1)
data_y = data1['satisfaction']
```

```python
# Esto lo dejamos comentado porque nos interesa el nombre de la columna

"""
data_x = data_x.values
data_y = data_y.values
"""
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.3)
```

```python
# Creamos el modelo predictivo

from sklearn.ensemble import RandomForestClassifier 

# Creamos 5000 decision trees
rf = RandomForestClassifier(n_estimators = 5000, random_state = 19)
```

```python
rf.fit(x_train, y_train)
```

```python
import pickle
```

```python
# Guardar en el disco
filename = 'model/rf.pkl'
pickle.dump(rf, open(filename, 'wb')) # rf = nuestro modelo
```
